```python
# combining columns to detect duplicates
# not able to add with Nan. so replacing with 'NA

crimes.fillna('NA', inplace=True)
crimes['totalnumbervictims_str'] = crimes['totalnumbervictims'].astype('str')
crimes['totalsuspects_str'] = crimes['totalsuspects'].astype('str')
crimes['combined'] = crimes['instanceid']+crimes['closed']+crimes['opening']+crimes['dayofweek']+crimes['victim_gender']+crimes['totalnumbervictims_str']+crimes['totalsuspects_str']
crimes[crimes['combined'].duplicated()]

# Below 4507 records can be removed as they are duplicates
```

```python
import matplotlib.pyplot as plt

plt.hist(crimes['totalsuspects'].replace('NA', np.nan))
```

```python
# calculated outliers after removing na values

y = crimes['totalsuspects'].replace('NA', np.nan)
y = y.dropna()
yq1, yq3 = np.percentile(y, [25 ,75])
yiqr = yq3-yq1

y = list(y)
y_outliers = 0
for i in y:
    if (i > (yq3 + 1.5 * yiqr) or i < (yq1 - 1.5 * yiqr)):
        y_outliers = y_outliers + 1

y_outliers
```

```python
x = crimes['totalnumbervictims'].replace('NA', np.nan)
x = x.dropna()
xq1, xq3 = np.percentile(x, [25 ,75])
xiqr = xq3-xq1

x = list(x)
x_outliers = 0
for i in x:
    if (i > (xq3 + 1.5 * xiqr) or i < (xq1 - 1.5 * xiqr)):
        x_outliers = x_outliers + 1

x_outliers
```
