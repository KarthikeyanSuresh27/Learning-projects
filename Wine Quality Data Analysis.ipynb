## Importing Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

winequality_red = pd.read_csv('winequality-red.csv', sep=';')
winequality_red
```

## What is the sample size? 
Ans: The data has 1599 observations and 12 attributes

```python
winequality_red.shape
```

## Any outliers? Do you have any concerns about the data quality?

Ans: There are no missing/null values in the dataset. However, there are several outliers.
Outliers are shown using Box plots below. 


```python
for column in winequality_red:
    plt.figure()
    winequality_red.boxplot([column])
```

```python
# Checking null values

winequality_red.isna().sum()
```

## How can you summarize the data of each variable in a concise way? What statistics are you going to present?

Ans: Summary statistics is used to represent data in consise way.
It shows count, mean, standard deviation, min, max, 25th percentile, 50th percentile and 75th percentile respectively

```python
winequality_red.describe()
```

## How can you visualize the distribution of each variable?
Ans: To understand the distribution, each variable can be visualized using histogram


```python
f, axes = plt.subplots(3, 4, figsize=(30, 20))
for i, ax in enumerate(axes.flatten()):
      sns.histplot(winequality_red.iloc[:, i], bins=20, kde=True, ax=ax)

```

## Do you see any skewed distributions?
Ans: Yes, there is skeweness in distribution. Below is the skewness measure.
Only Density and pH are close to normal distribution. All the other attributes are skewed to the right
            

```python
winequality_red.skew()
```
